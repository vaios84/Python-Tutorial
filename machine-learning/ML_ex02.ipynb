```python
import pandas as pd
from sklearn.tree import DecisionTreeClassifier

music_data = pd.read_csv('music.csv')
music_data
```

```python
x = music_data.drop(columns=['genre'])
x
```

```python
y = music_data['genre']
y
```

```python
model = DecisionTreeClassifier()
model.fit(x,y)
predictions = model.predict([ [26,1], [32,0] ])
predictions
```
